```python
result = pipeline.ask("Why are people unhappy with credit cards?")
print(result['answer'])
```

```python
test_questions = [
    "Why are people unhappy with credit cards?",
    "What are the main problems with savings accounts?",
    "What issues do customers face with money transfers?",
    "What are the most common personal loan complaints?",
    "How do customers describe their experience with customer service?",
    "What fraud related issues are customers reporting?",
    "What billing problems do credit card customers face?",
]

evaluation_results = []

for question in test_questions:
    print(f"\nQ: {question}")
    print("-" * 50)
    result = pipeline.ask(question)
    print(result['answer'])
    print()
    
    evaluation_results.append({
        'question': question,
        'answer': result['answer'],
        'sources': result['sources']
    })

print("\nAll questions answered!")
```

```python
for i, eval_result in enumerate(evaluation_results):
    print(f"\nQuestion {i+1}: {eval_result['question']}")
    print(f"\nAnswer: {eval_result['answer']}")
    print(f"\nTop 2 Sources:")
    for j, source in enumerate(eval_result['sources'][:2]):
        print(f"  {j+1}. {source['metadata']['product_category']} | {source['metadata']['issue']}")
    print("\n" + "="*60)
```
